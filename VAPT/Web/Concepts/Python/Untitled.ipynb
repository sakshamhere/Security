```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import warnings
```

```python
for i in range(1,17):
    url = f'https://www.crownandcaliber.com/collections/rolex-watches?page={i}'

    data = requests.get(url)

    soup = BeautifulSoup(data.content,'html.parser')
    title = soup.findAll('div',{'class':'card-title'})
    name = soup.findAll('div',{'class':'card-subTitle'})
    price = soup.findAll('div',{'class':'card-barcode'})

    for i,j,k in zip(price,title,name):
        print(j.text,k.text,i.text)
```
